```python
import numpy as np
import pandas as pd
```

```python
recent = pd.read_csv('../data/recent_data.csv')
recent = recent[recent['PTS']!='PTS']
recent = recent.dropna()

# Make it look like historical data
recent['fran_id'] = recent['Visitor/Neutral'].str.split().str[-1]
recent['opp_fran'] = recent['Home/Neutral'].str.split().str[-1]
fn = {'76ers':'Sixers', 'Blazers':'Trailblazers'}
recent['fran_id'].replace(fn, inplace=True)
recent['opp_fran'].replace(fn, inplace=True)
recent['pts'] = recent['PTS']
recent['opp_pts'] = recent['PTS.1']
recent['game_location'] = 'A'

# Parse dates
recent['day'] = recent['Date'].str.split().str[2].str[:-1]
recent['month'] = recent['Date'].str.split().str[1]
recent['year'] = recent['Date'].str.split().str[-1]
months = {'Jan' : 1,'Feb' : 2,'Mar' : 3,'Apr' : 4,'May' : 5,'Jun' : 6,'Jul' : 7,'Aug' : 8,'Sep' : 9,'Oct' : 10,'Nov' : 11,'Dec' : 12}
recent['month'].replace(months,inplace=True)

recent = recent[['fran_id','pts','opp_fran','opp_pts','game_location','month','day','year']]
recent
```

```python
data = pd.read_csv('../data/elo_1946-2015.csv')

# The current state of the NBA has 30 franchises. 
new_data = data[data['year_id']>=1977]

# Get rid of duplicate rows and keep important variables
new_data = new_data[new_data['_iscopy'] == 1] [['fran_id','pts','opp_fran','opp_pts','game_location','date_game']]
new_data['month'] = new_data['date_game'].str.split('/',n=3).str[0]
new_data['day'] = new_data['date_game'].str.split('/',n=3).str[1]
new_data['year'] = new_data['date_game'].str.split('/',n=3).str[2]
del new_data['date_game']
new_data
```

```python
final = pd.concat([recent,new_data])
final
```

```python
final.to_csv('historical_data.csv')
```
